#### <b>Data description</b>
<b>world_rank</b> - rank in international statistics<br>
<b>institution / school_name / university_name</b> - name of educational institution<br>
<b>country</b> - location of institution<br>
<b>national_rank</b> - local rank, place within its country<br>
<b>quality_of_education</b> - rank for quality of education<br>
<b>alumni_employment</b> - rank for alumni employment<br>
<b>quality_of_faculty</b> - rank for quality of faculty<br>
<b>publications</b> - rank for publications<br>
<b>influence</b> -rank for influence<br>
<b>citations</b> - number of students at the university<br>
<b>broad_impact</b> - rank for broad impact (only available for 2014 and 2015)<br>
<b>patents</b> - rank for patents<br>
<b>score</b> - total score, used for determining world rank<br>
<b>year</b> - year of ranking (2012 to 2015)<br>

<b>teaching</b> - university score for teaching (the learning environment)<br>
<b>international</b> - university score international outlook (staff, students, research)<br>
<b>research</b> - university score for research (volume, income and reputation)<br>
<b>citations</b> - university score for citations (research influence)<br>
<b>income</b> - university score for industry income (knowledge transfer)<br>
<b>total_score</b> - total score for university, used to determine rank<br>
<b>num_students</b> - number of students at the university<br>
<b>student_staff_ratio</b> - number of students divided by number of staff<br>
<b>international_students</b> - percentage of students who are international<br>
<b>female_male_ratio</b> - female student to Male student ratio<br>
<b>year</b> - year of the ranking (2011 to 2016 included)<br>

<b>alumni</b> - Alumni Score, based on the number of alumni of an institution winning nobel prizes and fields medals<br>
<b>award</b> - Award Score, based on the number of staff of an institution winning Nobel Prizes in Physics, Chemistry, Medicine, and Economics and Fields Medals in Mathematics<br>
<b>hici</b> - HiCi Score, based on the number of Highly Cited Researchers selected by Thomson Reuters<br>
<b>ns</b> - N&S Score, based on the number of papers published in Nature and Science<br>
<b>pub</b> - PUB Score, based on total number of papers indexed in the Science Citation Index-Expanded and Social Science Citation Index<br>
<b>pcp</b> - PCP Score, the weighted scores of the above five indicators divided by the number of full time academic staff<br>
<b>year</b> - year of ranking (2005 to 2015)<br>

<b>Private</b> - a factor with levels No and Yes indicating private or public university<br>
<b>Apps</b> - number of applications received<br>
<b>Accept</b> - number of applications accepted<br>
<b>Enroll</b> - number of new students enrolled<br>
<b>Top10perc</b> - new students from top 10% of H.S. class<br>
<b>Top25perc</b> - new students from top 25% of H.S. class<br>
<b>F.Undergrad</b> - number of fulltime undergraduates<br>
<b>P.Undergrad</b> - number of parttime undergraduates<br>
<b>Outstate</b> - Out-of-state tuition<br>
<b>Room.Board</b> - Room and board costs<br>
<b>Books</b> - Estimated book costs<br>
<b>Personal</b> - Estimated personal spending<br>
<b>PhD</b> - Pct. of faculty with Ph.D.’s<br>
<b>Terminal</b> - Pct. of faculty with terminal degree<br>
<b>S.F.Ratio</b> - Student/faculty ratio<br>
<b>perc.alumni</b> - Pct. alumni who donate
<b>Expend</b> - Instructional expenditure per student
<b>Grad.Rate</b> - Graduation rate<br>

<b>institute_type</b> - a grouping of institutions according to provider type (All Institutes (including preprimary education and subsidies to households, not separately shown), Elementary and secondary institutions (excludes preprimary), Higher education institutions).

```python
country_with_alotof_institution = country.groupby('country')['school_name'].count().sort_values(ascending = False)
country_with_alotof_institution.plot(figsize = (15, 6), grid = True, kind = 'bar')
plt.title("Number of possible educational institutions in each country")
plt.xlabel("Country names")
plt.ylabel("Number of institutions")
plt.show()
```

The following graph shows that there are a large number of high school graduate universities in the United States.

```python
school2['percentage'] = school2['enroll'] / school2['apps'] * 100
pd.options.display.float_format = '{:,.2f}'.format
school2.sort_values(by = ['percentage'], inplace = True, ascending = False)
school2 = school2.rename(columns = {'unnamed: 0': 'name'}, inplace = False)
```

Top 10 Universities with Great Admission Chances

```python
top = school2.head(10)
display(top[['name', 'percentage']])
```

Statistics showing how many new students at universities with high chances of admission entered from the top 10 and 25% H.S. Class

```python
sns.catplot(x = 'percentage', y = 'top10perc', data = top, kind = 'bar', aspect = 2);
sns.catplot(x = 'percentage', y = 'top25perc', data = top, kind = 'bar', aspect = 2);
plt.tight_layout();
plt.show()
```
